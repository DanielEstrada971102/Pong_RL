```python
import os
from PongGame_Env import *

from stable_baselines3 import A2C
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
log_path = 'Training/Logs'
Model_path = os.path.join('Training','Saved_Models', 'AC2_Pongenv')
```

```python
env = PongGame(game_speed = 2)

states = env.action_space.shape
actions = env.action_space.n
```

```python
episodes = 10
for episode in range(1, episodes+1):
    state = env.reset()
    done = False

    while not done:
        env.render(mode ='human')
        action = random.choice([0, 1, 2])
        n_state, reward, done, info = env.step(action)
    print('Episode:{} SCORE: {} Reward:{}'.format(episode, 
                                                  env.CURRENT_GAME_SCORE,reward))
env.close()
```

```python
env = PongGame(2)
episodes = 10
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    
    while not done:
        env.render(mode ='')
        action = random.choice([0, 1, 2])
        n_state, reward, done, info = env.step(action)
    print('Episode:{} SCORE: {} Reward:{}'.format(episode, 
                                                  env.CURRENT_GAME_SCORE,reward))

env.close()
```

```python
env = DummyVecEnv([lambda: env])
model = A2C("MlpPolicy", env, tensorboard_log=log_path, verbose=2)
```

```python
model.learn(total_timesteps=80000,  tb_log_name='A2C_Pongenv')
```

```python
model.save(Model_path)
```

```python
del model # remove to demonstrate saving and loading
```

```python
model = A2C.load(Model_path)
```

```python
env = PongGame(2)
#model = A2C.load(Model_path)

for episode in range(15):
    obs = env.reset()
    done = 0
    
    while not done:
        action, _ = model.predict(obs)
        obs, reward, done, info = env.step(action)
        env.render('human')
    
    print('Episode:{} Reward:{}'.format(episode, reward))

env.close()
```
